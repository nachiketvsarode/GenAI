```python
## Why functions?
num=24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
    
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
        
```

```python
## call this function 

even_or_odd(24)
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
### Default Parameters
def greet(name):
    print(f"Hello {name} welcome to the paradise")

greet("nachiket")
```

```python
### Default Parameters
def greet(name="Guest"): ###This is default paramter
    print(f"Hello {name} welcome to the paradise")

greet() ### not passing any parameter
```

```python

## Positional and keywords arguments

def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,67,8,9,2,3,4,5,64,54,5,5,6,"Nachiket")
```

```python
## Keyword  arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="nachiket", city="yyz", number="123456")
```

```python

def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional argument :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,"nachiket", name="nachiket", city="yyz", number="123456")
```
