The map() Function in Python

The map() funtion applies a given function to all items in a input list (or any other iterable) and rturns a map object (an iterator). This is particularly for transforming data in list comprehensively


```python
def square(x):
    return x*x

square(10)
```

```python
numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))
```

```python
## Lambda function with map
numbers=[1,2,3,4,5,6,7,8]
list(map(lambda x:x*x, numbers))
```

```python
## Map multiple iterable


numbers1=[1,2,3]
numbers2=[2,4,6]

added_numbers=list(map(lambda x,y:x+y, numbers1,numbers2))
print(added_numbers)
```

```python
## map() to to convert a list of strings to inters 

# use map to convert strings to int 

str_numbers = ['1', '2', '3', '4']
int_numbers = list(map(int, str_numbers))
print(int_numbers) 
```

```python
words =['apple', 'banana', 'cherry']
upper_word=list(map(str.upper,words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people=[
    {'name':'krish', 'age':32},
    {'name':'Jack', 'age':33}
]

list(map(get_name,people))

```
