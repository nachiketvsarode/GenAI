```python
## Why functions?
num=24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
    
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
        
```

```python
## call this function 

even_or_odd(24)
```

```python
## function with multiple parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
### Default Parameters
def greet(name):
    print(f"Hello {name} welcome to the paradise")

greet("nachiket")
```

```python
### Default Parameters
def greet(name="Guest"): ###This is default paramter
    print(f"Hello {name} welcome to the paradise")

greet() ### not passing any parameter
```
